```python
import pandas as pd
import os
```

```python
input_folder = r'D:\Dataset_Project\dataset_fyp\Network_dataset_preprocess'
#  Process the CSV file
for filename in os.listdir(input_folder):
    if filename.endswith('csv'):
        file_path = os.path.join(input_folder,filename)
        df = pd.read_csv(file_path)
        df.drop(columns=['ts', 'service','missed_bytes','dns_query',
                         'dns_qclass','dns_rcode','dns_AA','dns_RD','dns_RA','dns_rejected',
                         'ssl_version','ssl_cipher','ssl_resumed','ssl_established','ssl_subject','ssl_issuer',
                         'http_trans_depth','http_method','http_uri','http_referrer','http_version','http_request_body_len',
                         'http_response_body_len','http_status_code','http_user_agent','http_orig_mime_types','http_resp_mime_types',
                         'weird_name','weird_addl','weird_notice','type'], inplace=True)
        
        #Define the output path
        output_folder = r'D:\Dataset_Project\dataset_fyp\10Features'
        output_path = os.path.join(output_folder,f'10Features_{filename}')
        df.to_csv(output_path,index=False)
        print(f"Saved: {output_path}")
```

```python
# Input and output paths
input_file = r'D:\Dataset_Project\dataset_fyp\Network_dataset_preprocess\preprocess_Network_dataset_23.csv'
output_file = r'D:\Dataset_Project\dataset_fyp\10Features\t1_Network_dataset_23.csv'

# Read the CSV
df = pd.read_csv(input_file)
df.drop(columns=['ts', 'service','missed_bytes','dns_query',
                         'dns_qclass','dns_rcode','dns_AA','dns_RD','dns_RA','dns_rejected',
                         'ssl_version','ssl_cipher','ssl_resumed','ssl_established','ssl_subject','ssl_issuer',
                         'http_trans_depth','http_method','http_uri','http_referrer','http_version','http_request_body_len',
                         'http_response_body_len','http_status_code','http_user_agent','http_orig_mime_types','http_resp_mime_types',
                         'weird_name','weird_addl','weird_notice','type'], inplace=True)
        
# Save the processed file
df.to_csv(output_file, index=False)
print(f"Saved: {output_file}")
```
