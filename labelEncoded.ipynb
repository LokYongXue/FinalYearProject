```python
import pandas as pd
import os
from sklearn.preprocessing import LabelEncoder
```

```python
input_folder = r'D:\Dataset_Project\dataset_fyp\Network_dataset_preprocess'
for filename in os.listdir(input_folder):
    if filename.endswith('.csv'):
        file_path = os.path.join(input_folder,filename)
        df = pd.read_csv(file_path)
        # Getting the list of columns
        col = df.columns
        # Printing Number of columns
        print(f'{filename} : Number of columns :', len(col))
```

```python
input_folder = r'D:\Dataset_Project\dataset_fyp\Processed_Network_dataset\Processed_Network_dataset'
#  Process the CSV file
for filename in os.listdir(input_folder):
    if filename.endswith('csv'):
        file_path = os.path.join(input_folder,filename)
        df = pd.read_csv(file_path)
        df.drop(columns=['src_port', 'src_ip', 'dst_port','dst_ip'], inplace=True)
        
        #  Replace '-' with NaN
        df.replace('-', pd.NA, inplace=True)
        #  Handle missing values
        for col in df.columns:
            if df[col].dtype == 'object':
                # If column is string type, fill NaN with the most frequent value (mode) or 'unknown'
                if not df[col].mode().empty:
                    df[col] = df[col].fillna(df[col].mode()[0]) 
                else:
                    df[col].fillna('unknown', inplace=True)
            else:
                # If column is numeric, fill NaN with median
                df[col] = pd.to_numeric(df[col], errors='coerce')  # just in case strings are in numeric columns
                df[col] = df[col].fillna(df[col].median())
            
        label_encoders = {}

        for col in df.columns:
            if df[col].dtype == 'object':
                df[col] = df[col].astype(str)  # 🔧 Ensure consistent string format
                le = LabelEncoder()
                df[col] = le.fit_transform(df[col])
                label_encoders[col] = le

        #Define the output path
        output_folder = r'D:\Dataset_Project\dataset_fyp\Network_dataset_preprocess'
        output_path = os.path.join(output_folder,f'preprocess_{filename}')
        df.to_csv(output_path,index=False)
        print(f"Saved: {output_path}")
```
