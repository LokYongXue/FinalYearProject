```python
import pandas as pd
import os
from sklearn.preprocessing import LabelEncoder


input_folder = r'D:\Dataset_Project\dataset_fyp\10Features_multiclass'
output_folder = r'D:\Dataset_Project\dataset_fyp\10Features_multiclass_process'

all_data = []

print("=== Loading and preprocessing each file ===")

for filename in os.listdir(input_folder):
    if filename.endswith('.csv'):
        file_path = os.path.join(input_folder, filename)
        print(f"Processing: {filename}")

        df = pd.read_csv(file_path, low_memory=False)

         #  Replace '-' with NaN
        df.replace('-', pd.NA, inplace=True)

        #  Handle missing values
        for col in df.columns:
            if df[col].dtype == 'object':
                # If column is string type, fill NaN with the most frequent value (mode) or 'unknown'
                if not df[col].mode().empty:
                    df[col] = df[col].fillna(df[col].mode()[0]) 
                else:
                    df[col].fillna('unknown', inplace=True)
            else:
                # If column is numeric, fill NaN with median
                df[col] = pd.to_numeric(df[col], errors='coerce')  # just in case strings are in numeric columns
                df[col] = df[col].fillna(df[col].median())
            
        # Encode features (all object columns except label)
        for col in df.columns:
            if col != 'type' and df[col].dtype == 'object':
                df[col] = df[col].astype(str)
                le = LabelEncoder()
                df[col] = le.fit_transform(df[col])

        # Append for merging later
        all_data.append(df)

# =======================
# Merge all files
# =======================
print("\n=== Merging all files ===")
df = pd.concat(all_data, axis=0, ignore_index=True)
print(f"Merged dataset shape: {df.shape}")

# =======================
# Encode multiclass label 'type' once
# =======================
label_encoder = LabelEncoder()
df['type'] = df['type'].astype(str)
df['type'] = label_encoder.fit_transform(df['type'])
print("Classes found:", list(label_encoder.classes_))

# =======================
# Save merged preprocessed dataset
# =======================
output_path = os.path.join(output_folder, 'preprocess_merged_multiclass.csv')
df.to_csv(output_path, index=False)
print(f"Merged preprocessed dataset saved to: {output_path}")
```
